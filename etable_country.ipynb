### Function Descriptions

#### 1. `country_cause(file_path, sex_name, metric_name, year_start, year_end, save_path, Country_list)`
- **Purpose:** Creates age-standardized tables for each cause across different Sub-Saharan African countries.
- **Parameters:**
  - `file_path`: The path to the CSV file containing the data.
  - `sex_name`: The gender filter for the data (default: 'Both').
  - `metric_name`: The metric to filter the data (default: 'Rate').
  - `year_start`: The start year for comparison (default: 1990).
  - `year_end`: The end year for comparison (default: 2021).
  - `save_path`: Optional path to save the resulting tables as CSV files for each cause.
  - `Country_list`: A list of countries to include in the analysis (default: None, which includes all countries in the dataset).
- **Return:** A dictionary containing tables for each `cause_name` with rows representing countries and columns showing values for start year, end year, and percentage changes, formatted as `val (lower - upper)`.

#### 2. `std_age_cause(file_path, sex_name, metric_name, location_name, year_start, year_end, save_path)`
- **Purpose:** Creates age-standardized tables for different age groups for each cause at a specified location.
- **Parameters:**
  - `file_path`: The path to the CSV file containing the data.
  - `sex_name`: The gender filter for the data (default: 'Both').
  - `metric_name`: The metric to filter the data (default: 'Rate').
  - `location_name`: The name of the location to filter the data (default: 'Sub-Saharan Africa').
  - `year_start`: The start year for comparison (default: 1990).
  - `year_end`: The end year for comparison (default: 2021).
  - `save_path`: Optional path to save the resulting tables as CSV files for each cause.
- **Return:** A dictionary containing tables for each `cause_name`, with rows representing age groups ('15-49', '50-74', '75 up', and 'all age'), and columns showing values for start year, end year, and percentage change, formatted as `val (lower - upper)`.

#### 3. `std_location_measure(file_path, sex_name, metric_name, year_start, year_end, save_path, Country_list)`
- **Purpose:** Creates tables showing the standardized measure changes across specified countries and age groups.
- **Parameters:**
  - `file_path`: The path to the CSV file containing the data.
  - `sex_name`: The gender filter for the data (default: 'Both').
  - `metric_name`: The metric to filter the data (default: 'Rate').
  - `year_start`: The start year for comparison (default: 1990).
  - `year_end`: The end year for comparison (default: 2021).
  - `save_path`: Optional path to save the resulting tables as CSV files for each measure.
  - `Country_list`: A list of countries to include in the analysis (default: None, which includes all countries in the dataset).
- **Return:** A dictionary containing tables for each `measure_name`, with rows representing countries and columns representing percentage changes for different age groups ('14-49', '50-74', and '75+').

#### 4. `plot_heap(file_path, shape_file_path, num)`
- **Purpose:** Generates and plots maps showing the incidence change for different age groups and measures across Sub-Saharan African countries for a specified cause.
- **Parameters:**
  - `file_path`: The path to the CSV file containing the data.
  - `shape_file_path`: The path to the shapefile containing geographic data for the countries.
  - `num`: The index of the cause to plot from the unique causes in the dataset.
- **Return:** Displays multiple maps visualizing the percentage change in incidence for different age groups ('15-49', '50-74', '75+') and measures, for the selected cause, using the provided shapefile data.
