### Function Descriptions

#### 1. `Create_cause_measure_table(file_path, location_name, age_name, year_start, year_end, save_path)`
- **Purpose:** Creates a cause-measure table from a CSV file filtered by location, age group, and year range.
- **Parameters:**
  - `file_path`: The path to the CSV file containing the data.
  - `location_name`: The name of the location to filter the data (default: 'Sub-Saharan Africa').
  - `age_name`: The age group to filter the data (default: 'All ages').
  - `year_start`: The start year for the analysis (default: 1990).
  - `year_end`: The end year for the analysis (default: 2021).
  - `save_path`: Optional path to save the resulting table as a CSV file.
- **Return:** A table with rows representing different `cause_name` and `sex_name` combinations, and columns representing measures for the start year, end year, and percentage change, formatted as `val (lower - upper)`.

#### 2. `Create_std_cause_measure_table(file_path, location_name, age_name, year_start, year_end, save_path)`
- **Purpose:** Generates an age-standardized cause-measure table.
- **Parameters:**
  - `file_path`: The path to the CSV file containing the data.
  - `location_name`: The name of the location to filter the data (default: 'Sub-Saharan Africa').
  - `age_name`: The age group to filter the data (default: 'All ages').
  - `year_start`: The start year for the analysis (default: 1990).
  - `year_end`: The end year for the analysis (default: 2021).
  - `save_path`: Optional path to save the resulting table as a CSV file.
- **Return:** A table with age-standardized values for each `cause_name` and `sex_name`, showing metrics for the specified start and end years along with the percentage change, formatted as `val (lower - upper)`.

#### 3. `Create_std_year_measure_figure(file_path, location_name, sex_name, metric_name)`
- **Purpose:** Creates age-standardized trend figures for different causes and measures.
- **Parameters:**
  - `file_path`: The path to the CSV file containing the data.
  - `location_name`: The name of the location to filter the data (default: 'Sub-Saharan Africa').
  - `sex_name`: The gender to filter the data ('Both', 'Male', 'Female').
  - `metric_name`: The metric to filter the data (default: 'Rate').
- **Return:** This function generates a visual representation (figure) rather than a table. It provides age-standardized trend plots for each `cause_name` and `measure_name` combination over the years.

#### 4. `calculate_full_statistics(figure_1_path, hdi_path, location_name, sex_name, metric_name, hdi_name)`
- **Purpose:** Calculates the Pearson correlation, p-value, and 95% confidence intervals between age-standardized values and HDI.
- **Parameters:**
  - `figure_1_path`: The path to the CSV file containing the age-standardized data.
  - `hdi_path`: The path to the Excel file containing the HDI data.
  - `location_name`: The name of the location to filter the data (default: 'Sub-Saharan Africa').
  - `sex_name`: The gender to filter the data (default: 'Both').
  - `metric_name`: The metric to filter the data (default: 'Rate').
  - `hdi_name`: The HDI indicator to use (default: 'Human Development Index (value)').
- **Return:** A table with rows representing different `cause_name` and `measure_name` combinations, and columns containing the Pearson correlation coefficient, p-value, and 95% confidence interval (lower and upper bounds).
